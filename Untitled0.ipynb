## The function ▶

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
h = (4*x**3 - 27*x**2 + 5*x - 32) / (30*x**5 - 13*x**4 + 50*x**3 - 286*x**2 - 299*x - 70)
```

```python

h_num = sp.lambdify(x, h, 'numpy')

x_vals = np.linspace(-5, 5, 400)

y_vals = h_num(x_vals)
y_vals[np.isinf(y_vals)] = np.nan
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Plot of h(x)')
plt.ylim(-50, 50)
plt.show()
```

The decomposed fraction:

```python
h_apart = sp.apart(h)
print(sp.pretty(h_apart))
```

Integrate using the decomposition:

```python
int_apart = sp.integrate(h_apart, x)
print("\nIntegral of decomposed form:")
print(sp.pretty(int_apart))
```

Integrate using the original function:

```python
int_orig = sp.integrate(h, x)
print("\nIntegral of original function:")
sp.pprint(int_orig)
```

Comparison of the intergrals:

```python
# Differentiate the integrated form
diff_int_apart = sp.diff(int_apart, x)

# Simplify the differentiated form and compare to the original function
comparison = sp.simplify(diff_int_apart - h)

print("\nDifference between differentiated integral and original function:")
print(sp.pretty(comparison))

if comparison == 0:
    print("\nThe differentiated integral is equal to the original function.")
else:
    print("\nThe differentiated integral is NOT equal to the original function.")
```

The curve of y=log base5(2)

```python
x_vals = np.linspace(0, 10, 400)
y_vals = np.log(x_vals) / np.log(5)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$y = \log_5(2)$', color='purple')
plt.title(r'Graph of $y = \log_5(2)$')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

The arc length:

```python
# Define y as a symbolic variable using the expression for log base 5 of x
y = sp.log(x) / sp.log(5)

y_prime = sp.diff(y, x)
arc_length = sp.integrate(sp.sqrt(1 + y_prime**2), (x, 1, 2))
print(f"\nArc length from x=1 to x=2 for y=log_5(2):")
sp.pprint(arc_length)
print(f"≈ {arc_length.evalf()}")
```
